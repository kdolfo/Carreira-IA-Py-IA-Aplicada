## Conectando a IA no Google Colab

```python
import os
from google.colab import userdata

os.environ["GEMINI_API_KEY"] = userdata.get('GEMINI_API_KEY')
```

```python
from google import genai

client = genai.Client()
```

```python
resposta = client.models.generate_content(model="gemini-2.5-flash", contents="O quê é a inteligência artificial?")
```

```python
print(resposta.text)
```

```python
n = 1

while n <= 10:
  print("*")
  n = n + 1
```

```python
chat = client.chats.create(model="gemini-2.5-flash")
```

```python
resposta = chat.send_message("O quê é machine learning?")
```

```python
print(resposta.text)
```

```python
chat.get_history()
```

```python
prompt = input("digite a sua pergunta: ")

while prompt != "sair":
  resposta = chat.send_message(prompt)
  print(resposta.text)
  print("\n")
  prompt = input("digite a sua pergunta: ")
```
